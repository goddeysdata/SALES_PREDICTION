### IMPORTING LIBRARIES 

```python
###IMPORTING LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
### IMPORTING DATASET
loc ="/content/SALES.txt"
Sale_df =pd.read_csv(loc,sep="\s+",header=None)
Sale_df
```

```python
### CHECKING THE SHAPE OF THE DATASET
Sale_df.shape
```

```python
### CHECKING THE FIRST FIVE ROWS OF THE DATASET
Sale_df.head()
```

```python
### NAMING THE COLUMNS
Sale_df.columns=["Sales","Advertising"]
Sale_df.head()
```

```python
###  DESCRIBING THE DATASET, STATISTICAL INFORMATION
Sale_df.describe()
```

```python
### ASSIGNING X AND Y TO VARIOUS COLUMNS OF THE DATSET
X = Sale_df["Sales"].values
Y = Sale_df["Advertising"].values
```

```python
### PLOTTING A SCATTER PLOT TO SHOW RELATIONS 
plt.scatter(X,Y, color='blue',label='Scatter Plot')
plt.title('Relatioship between Sales and advertising')
plt.xlabel('Sales')
plt.ylabel('Advertising')
plt.show()
```

```python
### RESHAPING THE X AND Y
X =X.reshape(-1, 1)
Y =Y.reshape(-1, 1)
```

```python
Y.shape
X.shape
```

### TRAINING THE DATASET USING MACHINE LEARNING MODEL(LINEAR REGRESSION)

```python
### TRAINING DATASET
x_train, x_test, y_train, y_test = train_test_split(X, Y,test_size=0.33, random_state=42)
```

```python
### SHOWING SHAPE OF TRAINED AND TESTED PORTION OF THE DATASET
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
### TRAINING DATASET 
Sale_df_lm =LinearRegression()
Sale_df_lm.fit(x_train,y_train)
y_pred =Sale_df_lm.predict(x_test)
y_pred
```

```python
### PLOTTING THE TRAIN MODEL ON A SCATTER PLOT AND THE PREDICTION LINE 
plt.scatter(X,Y, color='blue',label='Scatter Plot')
plt.plot(x_test,y_pred,color='red',linewidth=3,label='Lm Line')
plt.title('Relatioship between Sales and advertising')
plt.xlabel('Sales')
plt.ylabel('Advertising')
plt.show()
```
